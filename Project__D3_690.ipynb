## Step 6 - Analyzing how charges are affected based on age and smoker status
- We plan in looking into the **age** of an individual and check if an individual is a **smoker or not**.
As we learned in the previous step, age plays a role with the **insurance premium charges** that
an individual would pay and we also learned that *smokers* also pay a higher premium charge than *non-smokers*.
- In this step, we will be evaluating how a smoker or non-smokers age may play a role with charges. 

### Step 6.1
Below, we are taking all of the *smokers* and then **aggregating their average charge** based on their age. There are
multiple indiviuals who may be of the **same age** and some who may be of a different age altogether.

```
"""
We take all of those who are smokers (smoker = yes) and then evaluating their insurance charges.
We use aggregates finding average charges. We will be calculating
the mean charges of all smokers of each age.
"""
df3 = df.loc[df['smoker']=='yes']


df3 = df3[["age","charges"]].groupby('age').agg(
    Mean_Insurance_Charges=pd.NamedAgg("charges", "mean")
)
df3 = df3.reset_index()
df3.head()
```

### Step 6.1.1
We will be taking the data for those individuals who are young adults between the **ages** of **20-40 years of age** 
as I believe being an young adult may have a role in premium charges.

```
#We take all smokers between ages 20-40 and their respective average charges.

df3 = df3.loc[df3['age'].between(20, 40)]
df3 = df3.rename(columns={'Mean_Insurance_Charges': 'smok_avg_charge'})
df3.head()
```

### Step 6.2
Below, we are taking all of the *non-smokers* and then **aggregating their average charge** based on their age. 

```
"""
We take all of those who are smokers (smoker = no) and then evaluating their insurance charges.
We use aggregates finding average charges. We will be calculating
the mean charges of all smokers of each age.
"""

df4 = df.loc[df['smoker']=='no']


df4 = df4[["age","charges"]].groupby('age').agg(
    Mean_Insurance_Charges=pd.NamedAgg("charges", "mean")
)
df4 = df4.reset_index()
df4.head()
```

### Step 6.2.1
We will be taking the data for all *non-smokers* who are young adults between the **ages** of **20-40 years of age**.

```
#We take all non-smokers between ages 20-40 and their respective average charges.

df4 = df4.loc[df4['age'].between(20, 40)]
df4 = df4.rename(columns={'Mean_Insurance_Charges': 'nosmok_avg_charge'})
df4.head()
```

### Step 6.3
- For this step, we will be merging the average charges of all the individuals who may be **smokers and non-smokers**
and keep them in seperate columns and they will differ by age *(age20-40)*.
- I did this as I felt it would be easier to *compare* the average charges of individuals in both **categories** side
by side.

```
##We will be merging the two dataframes with smoker and non-smoker data based on the ages (Age 20-40)
merge_df = df3.merge(df4, on = ['age'])
merge_df = merge_df.round(2)
merge_df.head()
```

### Step 6.4
- In this step we will be working on *visualization* where we are doing a *bar chart* to **compare** the charges
of smokers and non-smokers and their age.
- The age group as we know is between ages **20-40**.  

```
"""
Data Visualization of a bar graph where we are comparing the individuals who are amokers and those who 
are non-smokers and then we compare their average charges by age. Individuals are between ages 20-40.
"""
fig, ax = plt.subplots(figsize=(24, 12))

bar_width = 0.40
ax.bar(merge_df["age"] - bar_width/2, merge_df["smok_avg_charge"], bar_width, color = 'blue')
ax.bar(merge_df["age"] + bar_width/2, merge_df["nosmok_avg_charge"], bar_width, color = 'orange')
ax.set_title("Smoking status and charges for young adults", fontsize=20)
ax.set_xlabel("Age", fontsize=20)
ax.set_ylabel("Charges", fontsize=20)

plt.rc('xtick', labelsize=20) 
plt.rc('ytick', labelsize=20) 

fig.show()
```

As we can see the **results above** of young adults, those who are smokers are paying a **significantly higiher** amount 
than those who are non-smokers. 

### Step 6.5
- In this *step* below, we will be taking all ages int he dataset and comparing the average charges for each of the
ages for **all smokers vs non-smokers**. 
- The point of this *step* is to understand how the charges are affected for all ages (18 - 60's) and not

Below, we are taking the **average charges** of all individuals that are *non-smokers*, grouping them by their *age*

### Step 7.1
Below, we are taking all of the *smokers* and then **aggregating their average charge** based on their region.

```
"""
We take all of those who are smokers (smoker = yes) for
each of the four regions and evaluate their average
insurance premiums based on the region.
"""
region_smok_df = df.loc[df['smoker']=='yes']

region_smok_df = region_smok_df[["region","charges"]].groupby('region').agg(
    Mean_Insurance_Charges=pd.NamedAgg("charges", "mean")
)
region_smok_df = region_smok_df.rename(columns={'Mean_Insurance_Charges': 'reg_smok_avg_charge'})

region_smok_df = region_smok_df.reset_index()
region_smok_df
```

## Step 8 - Analyzing how charges are affected based on the gender and smoker status of an individual.
- We plan in looking into the **gender** of an individuals residence and check if an individual is a **smoker or not**.
- In this step, we will be evaluating how a smoker or non-smokers *gender* may play a role with charges. 
- The four main regions are *northwest*, *northeast*, *southwest*, *southeast* of the United States.

### Step 8.1
Below, we are taking all of the *smokers* and then **aggregating their average charge** based on their *gender*.

```
#We will be calculating the average insurance charges for each gender of all of those who smoke. 

gender_smok_df = df.loc[df['smoker']=='yes']

gender_smok_df = gender_smok_df[["gender","charges"]].groupby('gender').agg(
    Mean_Insurance_Charges=pd.NamedAgg("charges", "mean")
)
gender_smok_df = gender_smok_df.rename(columns={'Mean_Insurance_Charges': 'gender_smok_avg_charge'})

gender_smok_df = gender_smok_df.reset_index()
gender_smok_df
```

### Step 8.2
Below, we are taking all of the *non=smokers* and then **aggregating their average charge** based on their *gender*.

```
#We will be calculating the average insurance charges for each gender of all of those who do not smoke. 

gender_no_smok_df = df.loc[df['smoker']=='no']

gender_no_smok_df = gender_no_smok_df[["gender","charges"]].groupby('gender').agg(
    Mean_Insurance_Charges=pd.NamedAgg("charges", "mean")
)
gender_no_smok_df = gender_no_smok_df.rename(columns={'Mean_Insurance_Charges': 'gender_no_smok_avg_charge'})

gender_no_smok_df = gender_no_smok_df.reset_index()
gender_no_smok_df
```

### Step 8.3
- For this step, we will be merging the average charges of all
individuals who may be **smokers and non-smokers**
and keep them in seperate columns for comparison 
purposes.
- We take average charges of individuals in each of the **gender**.

```
"""
We are merging the data for both of the dataframes with the average 
charges per gender for smokers and non-smokers.
"""

gender_merg_df = gender_smok_df.merge(gender_no_smok_df, on = ['gender'])
gender_merg_df = gender_merg_df.round(2)
gender_merg_df
```
